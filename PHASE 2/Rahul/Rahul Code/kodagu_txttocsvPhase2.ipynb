```python
import csv

def extract_to_csv(input_file_path, output_directory):
    # Open the input file for reading
    with open(input_file_path, 'r') as input_file:
        # Read all lines from the file
        lines = input_file.readlines()

        # Extract station name from line 12 (assuming the name is between ':' and ',')
        station_name = lines[11].split(':')[1].split(',')[0].strip()

        # Extract CSV headings from line 14
        csv_headings = lines[13].split()

        # Extract data from lines 16 to 84
        data_lines = lines[15:84]

        # Create a CSV file with the extracted data
        output_file_path = f'{output_directory}/{station_name}.csv'
        with open(output_file_path, 'w', newline='') as csv_file:
            csv_writer = csv.writer(csv_file)

            # Write the CSV headings
            csv_writer.writerow(['YEAR', 'MN'] + csv_headings[2:])

            # Write the data to the CSV file
            for data_line in data_lines:
                # Split the line into columns
                columns = data_line.split()

                # Write the relevant data to the CSV file
                csv_writer.writerow(columns[:2] + columns[2:])

    print(f"Data extracted and saved to {output_file_path}")

# Replace 'path/to/your/input.txt' with the actual file path
# Replace 'path/to/your/output_directory' with the desired output directory
extract_to_csv('/content/drive/MyDrive/Phase 2-10 DISTRICTS 2012 TO 2023/Rahul/kodagu.txt', '/content/drive/MyDrive/Phase 2-10 DISTRICTS 2012 TO 2023/Rahul')
```
